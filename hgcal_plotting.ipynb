## Polygon definitions below

```python
def hexagon3d2(center,radius):
    a = [center[0],center[1]+radius,center[2]]
    b = [center[0]+radius/2,center[1]+(np.sqrt(3.)/2)*radius/2,center[2]]
    c = [center[0]+radius/2,center[1]-(np.sqrt(3.)/2)*radius/2,center[2]]
    d = [center[0],center[1]-radius,center[2]]
    e = [center[0]-radius/2,center[1]-(np.sqrt(3.)/2)*radius/2,center[2]]
    f = [center[0]-radius/2,center[1]+(np.sqrt(3.)/2)*radius/2,center[2]]

    return np.array([a,b,c,d,e,f])
def hexagon3d(center,radius):
    a = [center[0]+radius,center[1],center[2]]
    d = [center[0]-radius,center[1],center[2]]
    b = [center[0]+radius/2,center[1]+(np.sqrt(3.)/2)*radius,center[2]]
    c = [center[0]-radius/2,center[1]+(np.sqrt(3.)/2)*radius,center[2]]
    e = [center[0]-radius/2,center[1]-(np.sqrt(3.)/2)*radius,center[2]]
    f = [center[0]+radius/2,center[1]-(np.sqrt(3.)/2)*radius,center[2]]
    return np.array([a,b,c,d,e,f])

def square3d(center,side):
    a = [center[0]+side/2,center[1]+side/2,center[2]]
    b = [center[0]+side/2,center[1]-side/2,center[2]]
    c = [center[0]-side/2,center[1]-side/2,center[2]]
    d = [center[0]-side/2,center[1]+side/2,center[2]]
    
    return np.array([a,b,c,d])
```

```python
import numpy as np

path = '/home/llr/cms/sghosh/allschoolfinalmaterial/'
piarr = np.load(path+"piarraypALL.npy")

evt = 3
iarray = piarr[piarr[:,0]==evt]
iarray = iarray[:,[1,2,3,6]]
uarray = iarray[iarray[:,3]>0.3]


from sklearn.neighbors import NearestNeighbors
import numpy as np

nbrs = NearestNeighbors(n_neighbors=6, algorithm='ball_tree').fit(uarray[:,[0,1,2]])
distances, indices =  nbrs.kneighbors(uarray[:,[0,1,2]])


maxen = np.max(uarray[:,2])   
```

## 3D Plots with matplotlib only

```python
##### for pions  
import matplotlib as mpl
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as a3
import matplotlib.colors as colors
import pylab as pl


from mpl_toolkits.mplot3d import Axes3D 
%matplotlib notebook

 



cmap = mpl.cm.get_cmap('Blues')
cmap2 = mpl.cm.get_cmap('Greens')
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

for j,i in enumerate(uarray):

    if i[2]> 427:
        vtx = np.array(square3d(i[:3],3))
        tri = a3.art3d.Poly3DCollection([vtx])
        tri.set_color('blue')
    else:
        if np.sqrt(i[0]**2 + i[1]**2)> 68.4:
            vtx = np.array(hexagon3d(i[:3],0.7))
            #tri.set_color(cmap(i[3]/maxen))

            tri = a3.art3d.Poly3DCollection([vtx])
            if i[2] > 361.:
                tri.set_color('green')
            elif i[2] > 427.:
                tri.set_color('black')
            else:
                tri.set_color('red')
        else:

            vtx = np.array(hexagon3d(i[:3],0.5))
            #tri.set_color(cmap2(i[3]/maxen))

            tri = a3.art3d.Poly3DCollection([vtx])
            if i[2] > 373.:
                tri.set_color('green')
            elif i[2] > 427.:
                tri.set_color('black')
            else:
                tri.set_color('blue')
            #tri.set_alpha(i[3]/maxen)
    tri.set_edgecolor('k')
    tri.set_alpha(0.5)
    ax.add_collection3d(tri)

xa,ya,za,ua,va,wa = [],[],[],[],[],[]

for i in indices:
    for v in range(len(i)):

        if v==0: continue

        mean_x, mean_y, mean_z = uarray[i[0],0],uarray[i[0],1],uarray[i[0],2] # <-- starting point of vector
        xa.append(mean_x)
        ya.append(mean_y)
        za.append(mean_z)
        ua.append(uarray[i[v],0] - mean_x)
        va.append(uarray[i[v],1] - mean_y)
        wa.append(uarray[i[v],2] - mean_z)

        #ax.quiver(
        #    mean_x, mean_y, mean_z, # <-- starting point of vector
        #    uarray[i[v],0] - mean_x, uarray[i[v],1] - mean_y, uarray[i[v],2] - mean_z)#, # <-- directions of vector
        #     color = 'black', alpha = .8, lw = 1
        # )


ax.quiver(xa, ya, za, ua, va, wa, lw = 0.5,linestyles='solid' ,arrow_length_ratio=0.01,color='black',alpha=0.2)
tarray = iarray[iarray[:,3]>0.7]        
ax.set_xlim3d(np.min(tarray[:,0]), np.max(tarray[:,0]))
ax.set_ylim3d(np.min(tarray[:,1]), np.max(tarray[:,1]))
ax.set_zlim3d(np.min(tarray[:,2]), np.max(tarray[:,2]))
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.grid(True)
plt.axis('off')
plt.show()

        
```
